```python
import yfinance as yf
```

```python
start_date = '2020-01-01'
end_date = '2021-11-22'

Apple = yf.download('AAPL', start_date, end_date)
Microsoft = yf.download('MSFT', start_date, end_date)
SaudiAramco = yf.download('2222.SR', start_date, end_date)
Google = yf.download('GOOG', start_date, end_date)
Amazon = yf.download('AMZN', start_date, end_date)
```

```python
Apple['A10'] = Apple['Open'].rolling(10).mean()
Apple['A30'] = Apple['Open'].rolling(30).mean()
Apple['A90'] = Apple['Open'].rolling(90).mean()
Apple['Open'].plot(figsize = (15,9))
Apple['A10'].plot()
Apple['A30'].plot()
Apple['A90'].plot()
```
